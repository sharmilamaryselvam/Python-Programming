##**Conditional Statement**

A conditional statement allows a program to **make decisions** based on whether a **condition** is **True or False**. It is the foundation of **decision-making**.

Think of it like real life:

- If it is raining, take an umbrella.
- If your age is 18 or more, you can vote.
- Otherwise, do something else.

image.png

There are **3 ways** to write conditional statements

1. If
2. If-else
3. elif

1. The **if** Statement:
          
   - The starting point. It checks one single condition.

   > *Example: "If it is raining, take an umbrella."*
   

2. The **else** Statement:

   - Use else when you want an alternative action if the condition is False.

    >*Example: "Otherwise, wear sunglasses."*

3. The **elif** (Else-If) Statement:

   - Use elif ("else if") to check multiple conditions.

    >*Example: "If it is snowing, wear a coat; else if it is windy, wear a jacket."*

**Real Example: Even or Odd number**

```python
num = 8

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### **Flow**
1. Calculate `num % 2`
2. If result is `0`, condition is `True`
3. Print `"Even"`
4. Otherwise print `"Odd"`

### **Comparison Operators**
These are the tools used to **compare values** within the statement:

- **`==`** (Equal to)

- **`!=`** (Not equal to)

- **`> / <`** (Greater than / Less than)

- **`>= / <=`** (Greater than or equal to / Less than or equal to)

### **Logical Operators**
To make complex decisions, you can **combine conditions** using **logical operators:**

- **`and`**: True if **both** conditions are met (e.g., `age > 18 and has_id == True`).

- **`or`**: True if **at least one** condition is met (e.g., `is_weekend or is_holiday`).

- **`not`**: **Reverses** the result (e.g., `not is_raining`).

**1. Check Positive, Negative, or Zero**

   Write a program that takes a number and prints:

   - "Positive" if greater than 0
   
   - "Negative" if less than 0
   
   - "Zero" if equal to 0

```python
number = int(input("Enter your number:"))

if number > 0:
  print("Positive")
elif number < 0:
  print("Negative")
else:
  print("Zero")
```

**5. Check Leap Year**

Take a year as input and determine whether it is a leap year.

Rule:

- Divisible by 400 → Leap Year
- Divisible by 4 but not by 100 → Leap Year
- Otherwise → Not a Leap Year

```python
year = int(input("Enter year: "))

if (year % 400) == 0:
  print("Leap Year")
elif (year % 4 == 0) and (year % 100 != 0):
  print("Leap Year")
else:
  print("Not Leap Year")
```
