## **Variables**

variable is a named pointer or reference that points to a specific object or value stored in your computer's memory.

They are like **labeled containers for storing data**, and Python makes it flexible and easy to **use without** worrying about **explicit types** called **dynamic typing**.

##### **1. Creating Variables**
A variable is **created automatically** the exact moment you **assign a value** to it using the **`=`** operator

**Example:**

```python
x = 10

print(x)
```

##### **2. Naming Rules:**

1. Must start with a letter (a–z, A–Z) or underscore _.

   - *Eg:* `age = 25, _age = 25`

2. Can contain letters, numbers, and underscores.
   - *Eg:* ` ag1e = 25 ,age123 = 25, my_age = 25`
3. Cannot be a Python keyword like if, for, while, etc.
4. Case-sensitive (age and Age are different).
   - *Eg:* `age = 25, Age = 25.5, AGE = 'Twenty five'` all are different

#####**3. Assigning Values:**

  There are two ways to assign value to variables.

  1. Assigning Single value
  2. Multiple Assignment

**Example:**

```python
# Single Assignment

age = 10
print(age)

# Multiple ASsignment

age = year_old = 13
print(year_old)

age, dob = 12,'10-19-2014'
print(dob)
```

#####**4. Important Characteristics**

1. *Dynamic Typing:* Python determines the data type automatically based on the assigned value. You can reassign a variable to a completely different type of data later without errors.

2. *Multiple Assignment:* You can assign values to multiple variables at once in a single line (e.g., x, y, z = 1, 2, 3).

3. *Object References:* Python variables store references to objects rather than the actual value data itself.

##### **5. Best Practices**

1. Use **descriptive** names: **age** is **better** than **a**.
2. **Avoid** using Python **keywords** as variable names.
3. Use **snake_case** for multi-word variable names: **student_name**.

## **Data Types**

#####**4 Main data types:**

   1. *Integer (int)*  - Stores whole numbers.

   2. *Float (float)* - Stores decimal (floating-point) numbers.

   3. *String (str)* - Stores text enclosed in quotes.

   4. *Boolean (bool)* - Stores either True or False.

#####**Sequence Type:**
   5. *List* - Stores multiple items in an ordered collection.
   6. *Tuple* - Like a list, but cannot be changed after creation.

#####**Non-Sequence Type:**
   7. *Set (set)* - Stores unique items only.
   8. *Dictionary (dict)* - Stores data as key-value pairs. (Mapping)

**Example:**

```python
# Integer
temperature = -10
print(f"Int: {temperature}")

# Float
price = 99.99
print(f"Float: {price}")

# String
name = "Anne"
print(f"String: {name}")

# Boolean
is_logged_in = False
print(f"Boolean: {is_logged_in}")

# List
fruits = ["apple",1,"banana", "orange",5.21,True]
print(f"List: {fruits} -->(Ordered)")

# Tuple
coordinates = (10, 20)
print(f"Tuple: {coordinates} --> (Ordered)")

# Set
colors = {"red", "green", "blue"}
print(f"Set: {colors} --> (UnOrdered)")

# Dictionary

person = {
    "Name": "Sharmila",
    "Age" : 28,
    "Experience" : 1.10,
    "Working" : False
}

print(f"Dictionary: {person}")
```

### **Checking a Data Type**

Use the **`type()`** function to check what type of data is stored.

**Example:**

```python
type(person)
```
